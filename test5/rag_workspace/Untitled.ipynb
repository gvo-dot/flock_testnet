```python
from llama_index.core import SimpleDirectoryReader
```

```python
input_dir="./data/farcaster_files/"
```

```python
# only load markdown files
required_exts = [".md"]

reader = SimpleDirectoryReader(
    input_dir=input_dir,
    required_exts=required_exts,
    recursive=True,
)
```

```python
docs = reader.load_data()
```
